```python
df.rename(columns={'product_name_lenght':'product_name_length', 'product_description_lenght':'product_description_length', 'month_year':'date','total_price':'revenue'},inplace=True)
```

```python
df['date'] = pd.to_datetime(df['date'], format="%m/%d/%y")
df['date'] = df['date'].dt.strftime('%Y-%m-%d')
```

```python
#Merging ps1,2,3
df['product_score'] = df[['ps1', 'ps2', 'ps3']].mean(axis=1)

```

```python
# Drop the original individual product score and competitor price columns
df = df.drop(columns=['ps1', 'ps2', 'ps3'])
```
